```python
# import the weak learner, which is CART in this case
from sklearn.tree import DecisionTreeClassifier as DTree
import numpy as np
import time


class AdaBoostClassifier():
    def __init__(self, base_estimator=None, n_estimators=100):
        self.base_estimator = DTree(max_depth = 1, random_state = 1)
        self.n_estimators = n_estimators
        
    def get_error_rate(self, pred, Y):
        return sum(pred != Y) / float(len(Y))
    
    def train(self, X_train, Y_train, X_test, Y_test):
        n_train, n_test = len(X_train), len(X_test)
        # Initialize weights
        self.w = np.ones(n_train) / n_train
        self.alphas = []
        pred_train = 0
        pred_test = 0

        for i in range(self.n_estimators):
            # Fit a classifier with the specific weights
            self.base_estimator.fit(X_train, Y_train, sample_weight = self.w)
            pred_train_i = self.base_estimator.predict(X_train)
            pred_test_i = self.base_estimator.predict(X_test)
            
            # Indicator function
            diff = pred_train_i != Y_train
            same = [1 if i else -1 for i in diff]
            diff = [1 if i else 0 for i in diff]
            
            # Error
            err_m = np.sum(self.w * diff)
            # Alpha
            alpha_m = 0.5 * np.log( (1 - err_m) / float(err_m))
            self.alphas.append(alpha_m)
            
            z = 2*np.sqrt(err_m*(1-err_m))
            
            # New weights
            self.w = self.w * np.exp(np.float_(same) * alpha_m) / z
            
            # Add to prediction
            pred_train += pred_train_i * alpha_m
            pred_test += pred_test_i * alpha_m

        pred_train = np.sign(pred_train)
        pred_test = np.sign(pred_test)

        # Return error rate in train and test set
        return sum(pred_train != Y_train) / float(len(Y_train)), \
               sum(pred_test != Y_test) / float(len(Y_test))
        
    def predict(self, X):
        prediction = []
        for i in X:
            i = i.reshape(1, i.shape[0])
            pred = self.base_estimator.predict(i)
            pred = np.sum(pred * self.alphas)
            pred = 1 if pred > 0 else -1
            prediction.append(pred);
        return np.array(prediction)
```

```python
from sklearn.datasets import make_hastie_10_2
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cross_validation import train_test_split


X, y = make_hastie_10_2()
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

def base_clf(X_train, Y_train, X_test, Y_test, clf):
    clf.fit(X_train,Y_train)
    pred_train = clf.predict(X_train)
    pred_test = clf.predict(X_test)
    return sum(pred_train != Y_train) / float(len(Y_train)), \
           sum(pred_test != Y_test) / float(len(Y_test)),

def plot_error_rate(er_train, er_test):
    df_error = pd.DataFrame([er_train, er_test]).T
    df_error.columns = ['Training', 'Test']
    plot1 = df_error.plot(linewidth = 3, figsize = (8,6),
            color = ['lightblue', 'darkblue'], grid = True)
    plot1.set_xlabel('Number of iterations', fontsize = 12)
    plot1.set_xticklabels(range(0,600,50))
    plot1.set_ylabel('Error rate', fontsize = 12)
    plot1.set_title('Error rate vs number of iterations', fontsize = 16)
    plt.axhline(y=er_test[0], linewidth=1, color = 'red', ls = 'dashed')    
    
# Fit a simple decision tree first
clf_tree = DTree(max_depth = 1, random_state = 1)
er_tree = base_clf(X_train, Y_train, X_test, Y_test, clf_tree)

# Fit Adaboost classifier using a decision tree as base estimator
# Test with different number of iterations
er_train, er_test = [er_tree[0]], [er_tree[1]]
x_range = range(10, 600, 10)
for i in x_range:    
    er_i = AdaBoostClassifier(n_estimators=i).train(X_train,Y_train, X_test, Y_test)
    er_train.append(er_i[0])
    er_test.append(er_i[1])
# Compare error rate vs number of iterations
plot_error_rate(er_train, er_test)
```
